```python
# 2. Function to return the output as "SameCount" or "Programming"
# SameCount -- Lower and Upper chars have same count
#  Programming -- Count of Lower and Upper case is not same

#  Example #1 : PyThOn -- 3  P T O

#                                      -- 3  y h n
#  Output : SameCount

#  Example #2 : PythoN  -- 2 P N
#                                       -- 4 y t h o
#  Output  : Programming


#x=input("enter input")
def count(x):
    ucount=0
    lcount=0
    for i in x:
        if ord(i)>=65 and ord(i)<=90:
            print(i,end=" ")
            ucount=ucount+1
        elif ord(i)>=97 and ord(i)<=122:
            print(i,end=" ")
            lcount=lcount+1
    if ucount==lcount:
        print("\n same count")
    else:
        print("\n programing")
x=input("enter input")
count(x)
```

```python
# 4. Function to generate N random numbers in given range
import random
n=int(input("enter starting value"))
k=int(input("enter ending value"))
num=10
def rand(n,k,num):
    res=[]
    for i in range(num):
        res.append(random.randint(n,k))
    return res

print(rand(n,k,num))
```
